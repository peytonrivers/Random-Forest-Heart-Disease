```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=3, min_weight_fraction_leaf=0.1, random_state=42)
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
residual = y_test - predict
```

```python
predict
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print("Accuracy Score is: ", accuracy_score(y_test, predict))
print("Classification Report is: ", classification_report(y_test, predict))
```

```python
results = X_test.copy()
results['Actual'] = y_test
results['Predicted'] = predict
results['Correct'] = (results['Actual'] == results['Predicted']).astype(int)
```

```python
sample_data = df.sample(500, random_state=42)
correct = results[results['Correct'] == 1]
incorrect = results[results['Correct'] == 0]
results['Correct'] = (results['Actual'] == results['Predicted']).astype(int)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# --- Left: Actual results ---
sns.scatterplot(
    data=results, 
    x='age', 
    y='chest pain type', 
    hue='Actual', 
    palette='coolwarm', 
    alpha=0.7, 
    ax=axes[0]
)
axes[0].set_title("Actual Chest Pain Status")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Chest Pain(1 = lowest, 4 = highest)")
axes[0].legend(title="Heart Disease (0=No, 1=Yes)")

# --- Right: Correct vs Incorrect ---
sns.scatterplot(
    data=results, 
    x='age', 
    y='chest pain type', 
    hue='Correct', 
    palette={1:'green', 0:'red'}, 
    alpha=0.7, 
    ax=axes[1]
)
axes[1].set_title("Model Prediction Accuracy")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Chest Pain(1 = lowest, 4 = highest)")
axes[1].legend(title="Prediction (Green=Correct, Red=Wrong)")

plt.tight_layout()
plt.show()
```

```python
importances = model.feature_importances_
feat_importances = pd.Series(importances, index=x.columns)
feat_importances.sort_values().plot(kind='barh', figsize=(10,6))
plt.title("Random Forest Feature Importance")
plt.show()
```
